# **INFORMASI DATASET**
## Latar Belakang

Dalam dunia bisnis ritel, pemahaman terhadap perilaku pelanggan menjadi salah satu faktor krusial dalam menentukan strategi pemasaran, peningkatan layanan, serta pengambilan keputusan berbasis data. Salah satu pendekatan yang umum digunakan untuk memahami perilaku pelanggan adalah melalui segmentasi, yaitu proses pengelompokan pelanggan ke dalam beberapa segmen atau kelompok berdasarkan karakteristik tertentu.

Dengan menerapkan metode **unsupervised learning**, khususnya **clustering**, proyek ini bertujuan untuk membentuk segmentasi pelanggan yang merepresentasikan pola-pola pembelian yang serupa. Hasil segmentasi ini diharapkan dapat memberikan insight yang bernilai bagi pihak manajemen, khususnya dalam menyusun strategi promosi yang lebih tepat sasaran, mengembangkan program loyalitas, serta mengoptimalkan layanan pelanggan.

**Exploratory Data Analysis (EDA)**

```python
print("Numerik vs Numerik → Pearson Correlation.")
print("Nominal vs Nominal → Cramer’s V.")
print("Numerik vs Nominal → ETA Correlation Ratio.")

numeric_cols = ['TransactionAmount', 'CustomerAge', 'TransactionDuration', 'LoginAttempts', 'AccountBalance']
nominal_cols = ['TransactionType', 'Channel', 'CustomerOccupation']

plt.style.use('seaborn-v0_8-white')
plt.rcParams['font.family'] = 'Arial'

assoc = associations(
    df[numeric_cols + nominal_cols],
    numerical_columns=numeric_cols,
    nominal_columns=nominal_cols,
    mark_columns=True,
    num_num_assoc='pearson',
    nom_nom_assoc='cramer',
    nom_num_assoc='correlation_ratio',
    plot=False  # We'll plot it separately to customize
)

# Now create the figure with our customizations
plt.figure(figsize=(14, 10))
ax = sns.heatmap(
    assoc['corr'],
    mask=np.triu(np.ones_like(assoc['corr'], dtype=bool)),
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    linewidths=1,
    square=True,
    cbar_kws={'shrink': 0.8},
    vmax=1, 
    vmin=-1
)

ax.set_title(
    'Heatmap Asosiasi Fitur Pelanggan Bank',
    fontsize=14, weight='bold', pad=20
)

plt.figtext(
    0.1, 1.01, 'Analisis asosiasi fitur numerik dan kategorikal pelanggan bank',
    fontsize=20, fontweight='bold', color='#b30000'
)
plt.figtext(
    0.1, 0.98, 'Bagaimana hubungan antara berbagai tipe fitur dalam dataset ini?',
    fontsize=16, color='black'
)
plt.figtext(
    0.5, 0.75, "Matriks asosiasi membantu memahami\nhubungan antar semua tipe fitur\ndalam satu visualisasi.",
    fontsize=14, ha='left', va='top',
    bbox=dict(facecolor='white', edgecolor='#e0e0e0', boxstyle='round,pad=0.7')
)

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
fig = plt.figure(figsize=(12, 3.5 * len(numeric_cols)))
gs = fig.add_gridspec(len(numeric_cols) + 1, 2, height_ratios=[0.1] + [1] * len(numeric_cols), hspace=0.5)

header_color, sub_color = '#b30000', '#333'
hist_color, box_color, med_color = sns.color_palette("Blues")[3], sns.color_palette("Reds")[3], '#c0392b'

ax_head = fig.add_subplot(gs[0, :]); ax_head.axis('off')
ax_head.text(
    0.5, 0.8, 'ANALISIS DISTRIBUSI FITUR NUMERIK',
    fontsize=18, weight='bold', color=header_color, ha='center', fontname='Arial'
)
ax_head.text(
    0.5, -0.5, 'Distribusi dan karakteristik statistik fitur transaksi bank',
    fontsize=13, color=sub_color, ha='center', fontname='Arial'
)

for i, col in enumerate(numeric_cols):
    ax1 = fig.add_subplot(gs[i+1, 0])
    sns.histplot(df[col], kde=True, ax=ax1, bins=15, color=hist_color, alpha=0.8)
    med = df[col].median()
    ax1.axvline(med, color=med_color, ls='--', lw=1.5)
    ax1.set_title(f'Distribusi {col}', pad=10, color=sub_color, fontname='Arial')
    ax1.annotate(
        f'Median: {med:.2f}', (med, ax1.get_ylim()[1]*0.9), xytext=(5,0),
        textcoords='offset points', color=med_color, fontname='Arial'
    )

    ax2 = fig.add_subplot(gs[i+1, 1])
    sns.boxplot(x=df[col], ax=ax2, color=box_color, width=0.4)
    ax2.set_title(f'Boxplot {col}', pad=10, color=sub_color, fontname='Arial')

plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn-v0_8-white')

header_color, sub_color = '#b30000', '#333'
palette = sns.color_palette("Blues_r", n_colors=12)
text_color = '#2d3436'
bg_color = 'white'

n_plots = len(nominal_cols)
fig = plt.figure(figsize=(4*n_plots, 6.5), facecolor=bg_color)

ax_head = fig.add_axes([0.1, 0.9, 0.8, 0.1])
ax_head.axis('off')
ax_head.text(
    0.5, 0.6, 'ANALISIS DISTRIBUSI FITUR KATEGORIK',
    fontsize=18, weight='bold', color=header_color, ha='center', fontname='Arial'
)
ax_head.text(
    0.5, 0.1, 'Distribusi fitur nominal/kategorik transaksi bank',
    fontsize=13, color=sub_color, ha='center', fontname='Arial'
)

axes = []
if n_plots > 1:
    axes = fig.subplots(1, n_plots)
else:
    axes = [fig.add_subplot(111)]

for i, col in enumerate(nominal_cols):
    ax = axes[i] if n_plots > 1 else axes[0]
    
    value_counts = df[col].value_counts(normalize=True).mul(100).sort_values(ascending=False)
    
    barplot = sns.barplot(
        x=value_counts.index,
        y=value_counts.values,
        ax=ax,
        palette=palette,
        orient='v',
        saturation=0.8
    )
    
    ax.set_facecolor(bg_color)
    ax.set_title(
        f'Distribusi {col}',
        fontsize=12, color=sub_color, 
        fontname='Arial', weight='bold',
        pad=12
    )
    ax.set_xlabel('')
    ax.set_ylabel('Persentase (%)', fontname='Arial', color=text_color, labelpad=10)
    
    max_percentage = value_counts.max()
    ax.set_ylim(0, min(100, max_percentage * 1.15))
    
    for p in barplot.patches:
        height = p.get_height()
        x = p.get_x() + p.get_width()/2
        
        if height > 10:
            label_y = height - 1
            va = 'top'
            color = 'white'
            weight = 'bold'
        else:
            label_y = height + 1
            va = 'bottom'
            color = text_color
            weight = 'normal'
            
        ax.text(
            x, label_y,
            f'{height:.1f}%',
            ha='center', va=va,
            color=color,
            fontsize=9,
            fontname='Arial',
            weight=weight
        )
    
    for spine in ['top', 'right']:
        ax.spines[spine].set_visible(False)
    ax.spines['left'].set_color('#ddd')
    ax.spines['bottom'].set_color('#ddd')
    ax.grid(axis='y', linestyle='--', alpha=0.3, color='#aaa')

plt.tight_layout(rect=[0, 0, 1, 0.9])
plt.show()
```

# **3. Pembersihan dan Pra Pemrosesan Data**

**Drop Kolom yang Tidak Digunakan**

```python
drop_cols = ['TransactionID', 'AccountID', 'TransactionDate', 'MerchantID',
             'Location', 'DeviceID', 'IP Address', 'PreviousTransactionDate']
df_cluster = df.drop(columns=drop_cols).copy()

numeric_cols = df_cluster.select_dtypes(include=['int64', 'float64']).columns.tolist()
nominal_cols = df_cluster.select_dtypes(include=['object']).columns.tolist()

print(f"Kolom Numerik             : {numeric_cols}")
print(f"Kolom Kategorik (Nominal) : {nominal_cols}")
```

**Duplikasi Data**

```python
print(df_cluster.shape)
print(f"Data duplikat: {df_cluster.duplicated().sum()}")
```

```python
df_cluster.drop_duplicates(keep='first', inplace=True)
print(df_cluster.shape)
```

**Handling Missing Value**

```python
df_cluster.isnull().sum()
```

```python
df_cluster.dropna(inplace=True)
df_cluster.isnull().sum()
```

**Encoding**

```python
df_cluster_encoded = pd.get_dummies(df_cluster, columns=nominal_cols, drop_first=False)
df_cluster_encoded.head()
```

**Scaling**

```python
scaler = StandardScaler()

df_cluster_encoded_scaled = df_cluster_encoded.copy()
df_cluster_encoded_scaled[numeric_cols] = scaler.fit_transform(df_cluster_encoded_scaled[numeric_cols])

df_cluster_encoded_scaled.head()
```

**Tidak dilakukan penanganan Outlier, karena akan menghilangkan informasi terutama anomali dalam data**

# **4. Membangun Model Clustering**

```python
df_cluster_encoded_scaled.describe()
```

**K-Means**

```python
model = KMeans(init="k-means++", random_state=42, n_init=10)
visualizer = KElbowVisualizer(model, k=(2, 10), metric='silhouette', timings=True)
visualizer.fit(df_cluster_encoded_scaled)
visualizer.show()
```

```python
optimal_k = visualizer.elbow_value_

kmeans = KMeans(n_clusters=optimal_k, init="k-means++", random_state=42, n_init=10)
kmeans.fit(df_cluster_encoded_scaled)
```

```python
cluster_labels = kmeans.labels_
```

```python
joblib.dump(kmeans, 'model_clustering.h5')
```

```python
score = silhouette_score(df_cluster_encoded_scaled, cluster_labels)
print(f"Silhouette Score: {score:.4f}")
```

```python
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(df_cluster_encoded_scaled)

df_umap = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])
df_umap['Cluster'] = cluster_labels.astype(str)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_umap, x='UMAP1', y='UMAP2', hue='Cluster', palette='tab10', s=60)
plt.title('Visualisasi Clustering dengan UMAP')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```

__Berdasarkan hasil eksperimen menggunakan K-Means++:__

- **K-Means++** pada seluruh fitur memberikan **Silhouette Score sebesar 0.2017** (k=3).

__Silhouette score yang tergolong rendah mengindikasikan bahwa hasil clustering belum optimal, kemungkinan disebabkan oleh:__

> - Banyaknya fitur yang tidak relevan, yang dapat menyebabkan tingginya dimensi dan meningkatkan noise dalam proses clustering.  
> - Keberadaan fitur dengan variansi rendah yang cenderung tidak memberikan informasi penting untuk pemisahan klaster.  

---

#### Solusi: Feature Selection untuk Meningkatkan Performa Clustering

__Untuk memperbaiki kualitas clustering, pendekatan feature selection dapat diterapkan melalui:__

- **Variance Threshold**  
  Mengeliminasi fitur dengan variansi rendah karena fitur-fitur ini cenderung tidak berkontribusi terhadap pemisahan data dalam klaster.

> Setelah proses feature selection, clustering dapat diulang untuk mengevaluasi apakah terjadi peningkatan pada Silhouette Score.

**K-Means + Feature Selection**

```python
selector = VarianceThreshold(0.2)
df_selected = selector.fit_transform(df_cluster_encoded)
selected_columns = df_cluster_encoded.columns[selector.get_support()]

selected_columns
```

```python
df_cluster_selected = pd.DataFrame(df_selected, columns=selected_columns)

scaler = StandardScaler()
df_cluster_selected_scaled = pd.DataFrame(scaler.fit_transform(df_cluster_selected), columns=selected_columns)
df_cluster_selected_scaled.head()
```

```python
model = KMeans(init="k-means++", random_state=42, n_init=10)
visualizer = KElbowVisualizer(model, k=(2, 10), metric='silhouette', timings=True)
visualizer.fit(df_cluster_selected_scaled)
visualizer.show()
```

```python
optimal_k = visualizer.elbow_value_

# Clustering
kmeans = KMeans(n_clusters=optimal_k, init="k-means++", random_state=42, n_init=10)
kmeans.fit(df_cluster_selected_scaled)
```

```python
cluster_labels = kmeans.labels_
```

```python
joblib.dump(kmeans, 'model_fs_clustering.h5')
```

```python
score = silhouette_score(df_cluster_selected_scaled, cluster_labels)
print(f"Silhouette Score: {score:.4f}")
```

```python
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(df_cluster_selected_scaled)

df_umap = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])
df_umap['Cluster'] = cluster_labels.astype(str)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_umap, x='UMAP1', y='UMAP2', hue='Cluster', palette='tab10', s=60)
plt.title('Visualisasi Clustering dengan UMAP')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```

**K-Means + Feature Selection + PCA**

```python
target_variance = 0.9
pca_full = PCA()
pca_full.fit(df_cluster_selected_scaled)

cumulative_variance = pca_full.explained_variance_ratio_.cumsum()
n_components_optimal = (cumulative_variance < target_variance).sum() + 1

pca = PCA(n_components=n_components_optimal, random_state=42)
df_pca_array = pca.fit_transform(df_cluster_selected_scaled)
df_pca = pd.DataFrame(df_pca_array, columns=[f'PC{i+1}' for i in range(n_components_optimal)])

print(f"Jumlah komponen PCA untuk menjelaskan {target_variance*100:.0f}% varians: {n_components_optimal}")
df_pca.head()
```

```python
model = KMeans(init="k-means++", random_state=42, n_init=10)
visualizer = KElbowVisualizer(model, k=(2, 10), metric='silhouette', timings=False)
visualizer.fit(df_pca)
visualizer.show()
```

```python
optimal_k = visualizer.elbow_value_

kmeans = KMeans(n_clusters=optimal_k, init="k-means++", random_state=42, n_init=10)
kmeans.fit(df_pca)
```

```python
cluster_labels = kmeans.labels_
```

```python
joblib.dump(kmeans, 'model_fs_pcs_clustering.h5')
```

```python
score = silhouette_score(df_pca, cluster_labels)
print(f"Silhouette Score: {score:.4f}")
```

```python
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(df_pca.drop(columns='cluster', errors='ignore'))

df_umap = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])
df_umap['Cluster'] = cluster_labels.astype(str)

centroids = df_umap.groupby('Cluster')[['UMAP1', 'UMAP2']].mean().reset_index()

palette = sns.color_palette("Set1", n_colors=df_umap['Cluster'].nunique())

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_umap,
    x='UMAP1', y='UMAP2',
    hue='Cluster',
    palette=palette,
    s=60,
    alpha=0.8,
    edgecolor='black'
)

plt.scatter(
    centroids['UMAP1'], centroids['UMAP2'],
    s=250,
    c='black',
    marker='X',
    label='Centroid'
)

for _, row in centroids.iterrows():
    plt.text(
        row['UMAP1'] + 0.3, row['UMAP2'],
        f"Cluster {row['Cluster']}",
        fontsize=10,
        weight='bold',
        color='black',
        bbox=dict(boxstyle="round,pad=0.2", edgecolor='black', facecolor='white', alpha=0.7)
    )

plt.title('Visualisasi Clustering dengan UMAP', fontsize=14)
plt.xlabel('UMAP1')
plt.ylabel('UMAP2')
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
