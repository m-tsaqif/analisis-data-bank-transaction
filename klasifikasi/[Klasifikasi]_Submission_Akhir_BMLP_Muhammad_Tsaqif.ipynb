# **1. Import Library**

```python
import time

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier

from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from sklearn.metrics import (
    classification_report,
    precision_score,
    recall_score,
    f1_score,
    accuracy_score,
    roc_auc_score
)
import joblib
```

# **2. Memuat Dataset dari Hasil Clustering**

```python
df = pd.read_csv("../datasets/data_clustering.csv")

print(df.shape)
df.head()
```

# **3. Data Splitting**

```python
X = df.drop(columns='Target', axis=1)
y = df.Target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# **4. Membangun Model Klasifikasi**

```python
categorical_pipeline = Pipeline([
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

categorical_columns = ['TransactionType', 'Channel', 'CustomerOccupation']

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_columns),
    ],
    remainder='passthrough',
    force_int_remainder_cols=False
)
```

```python
models = {
    "Decision Tree": DecisionTreeClassifier(
        max_depth=5,
        min_samples_leaf=10,
        random_state=42
    ),
    
    "Random Forest": RandomForestClassifier(
        n_estimators=100, 
        max_depth=10, 
        random_state=42
    ),
    
    "Gradient Boosting": GradientBoostingClassifier(
        n_estimators=100,
        learning_rate=0.1,
        max_depth=3,
        random_state=42
    ),
    
    "XGBoost": XGBClassifier(
        n_estimators=100,
        max_depth=4,
        learning_rate=0.1,
        eval_metric='mlogloss',
        random_state=42
    )
}
```

```python
# Stratified K-Fold Cross Validation (Mengatasi Imbalance Data)
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

for name, model in models.items():
    clf_pipeline = Pipeline([
        ("preprocessor", preprocessor),
        ("classifier", model)
    ])
    
    accuracy_scores = []
    precision_scores = []
    recall_scores = []
    f1_scores = []
    auc_scores = []

    start_time = time.time()
    
    for train_idx, val_idx in skf.split(X_train, y_train):
        X_train_fold, X_val_fold = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_train_fold, y_val_fold = y_train.iloc[train_idx], y_train.iloc[val_idx]
        
        clf_pipeline.fit(X_train_fold, y_train_fold)
        y_pred = clf_pipeline.predict(X_val_fold)
        y_proba = clf_pipeline.predict_proba(X_val_fold)
        
        accuracy_scores.append(accuracy_score(y_val_fold, y_pred))
        precision_scores.append(precision_score(y_val_fold, y_pred, average='weighted'))
        recall_scores.append(recall_score(y_val_fold, y_pred, average='weighted'))
        f1_scores.append(f1_score(y_val_fold, y_pred, average='weighted'))
        
        try:
            auc = roc_auc_score(y_val_fold, y_proba, multi_class='ovr', average='weighted')
            auc_scores.append(auc)
        except ValueError:
            auc_scores.append(np.nan)

    training_time = time.time() - start_time
    
    avg_accuracy = np.mean(accuracy_scores)
    avg_precision = np.mean(precision_scores)
    avg_recall = np.mean(recall_scores)
    avg_f1 = np.mean(f1_scores)
    avg_auc = np.nanmean(auc_scores)
    
    print(f"\n=== {name} ===")
    print(f"Akurasi rata-rata dari Cross-Validation: {avg_accuracy:.4f}")
    print(f"Presisi rata-rata dari Cross-Validation: {avg_precision:.4f}")
    print(f"Recall rata-rata dari Cross-Validation: {avg_recall:.4f}")
    print(f"F1-Score rata-rata dari Cross-Validation: {avg_f1:.4f}")
    print(f"ROC AUC (OvR) rata-rata dari Cross-Validation: {avg_auc:.4f}")
    print(f"Waktu pelatihan: {training_time:.4f} detik")
    
    joblib.dump(clf_pipeline, f'{name}_classification.h5')
```

**Model yang dilakukan Hyperparameter Tuning adalah `Decision Tree`, karena semua model memberikan hasil metrik yang bagus, namun `Decision Tree` membutuhkan waktu training yang lebih sedikit**

```python
dt_model = DecisionTreeClassifier(random_state=42)

param_grid = {
    'classifier__max_depth': [3, 5, 7, 10, None],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 5],
    'classifier__criterion': ['gini', 'entropy']
}

clf_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", dt_model)
])

grid_search = GridSearchCV(
    clf_pipeline,
    param_grid,
    cv=3,
    scoring='roc_auc_ovr',
    n_jobs=-1,
    verbose=1
)

start_time = time.time()

grid_search.fit(X_train, y_train)

training_time = time.time() - start_time
```

```python
print(f"Best parameters found: {grid_search.best_params_}")
```

```python
print(f"Best score from GridSearchCV: {grid_search.best_score_:.4f}")
```

```python
print(f"Best estimator from GridSearchCV:")
grid_search.best_estimator_
```

```python
y_pred = grid_search.predict(X_test)
print("\nLaporan Klasifikasi pada Data Test:")
print(classification_report(y_test, y_pred))
```

```python
print(f"Waktu pelatihan dengan GridSearchCV: {training_time:.4f} detik")
```

```python
cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=grid_search.classes_)
fig, ax = plt.subplots(figsize=(6, 6))
disp.plot(ax=ax, cmap='Blues', values_format='d', colorbar=False)
plt.title("Confusion Matrix (Decision Tree)")
plt.grid(False)
plt.show()
```
